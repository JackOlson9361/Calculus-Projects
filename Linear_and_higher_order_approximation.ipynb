# Approximating functions with derivatives

Functions can be approximated in several different ways.

A very simple way is to just choose some inputs that are spread out:

 $$  x = 1 \hspace{1cm} x = 3 \hspace{1cm} x = 5 \hspace{1cm} x = 7 \hspace{1cm} x = 9 $$

and then use the outputs that are associated with them:

In the case of $y = \sqrt{x}$, the outputs would be

$$ y = \sqrt{1} \hspace{1cm} y = \sqrt{3} \hspace{1cm} y = \sqrt{5} \hspace{1cm} y = \sqrt{7} \hspace{1cm} y = \sqrt{9}$$

In this scenario, we'd choose the best output by just finding the a nearby input.  So, to approximate the function when $x = 8.2$, we'd estimate that the output is 

$$ \sqrt{8.2} \approx \sqrt{9} = 3 $$
